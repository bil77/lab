#Functions

```python
def fitCPW2(f,S21_max,Q,f0):
    """ Complex fit using Lorentzian curve and error term.
    S21_max = Max value S21 linear 
    Q = loaded Quality Factor
    Amp0 = amplitude error
    Phase0 = Phase error
    """
    
    
    S21 = S21_max/(1+1j*Q*(f/f0-f0/f))   # f**3*(f*(f*floor)+floor2) #

    return S21

def back(x,a,b,c,d,e,f):
    """ Background fit, quaratic in phase
    AmpX = amplitude error
    PhaseX = Phase error
    """
    
    S21 = (a + 1j*b) + (c + 1j*d)*x + (e + 1j*f)*x**2 
    
    #a/(1+1j*b) + c/(1+1j*d/x) + e/(1+1j*f/x)
    
    # (a + 1j*b) + (c + 1j*d)*x + (e + 1j*f)*x**2  
    
    # a*exp(-1j*b) + c*exp(-1j*d)*x + e*exp(-1j*f)*x**2  
    
    # a + b*x + c*x**2 + d*x**3 + e*x**4 + f*x**5

    return S21

def residual_2(pars2, freq, S21):
        S21_max, Q, f0, Amp0, Phase0, Amp1,Phase1,Amp2,Phase2 = pars2
    #     
        diff =  S21 - ( fitCPW2(freq,S21_max,Q,f0) + back(freq,Amp0,Phase0,Amp1,Phase1,Amp2,Phase2) ) 
        return  (abs(diff))#diff.real**2 + diff.imag**2#    
```

# Rotate data complex plane

```python
x = Y.real
y = Y.imag

x_m = mean(x)
y_m = mean(y)

def calc_R(xc, yc):
    """ calculate the distance of each 2D points from the center (xc, yc) """
    return sqrt((x-xc)**2 + (y-yc)**2)


def f_3(beta, x):
    """ implicit definition of the circle """
    return (x[0]-beta[0])**2 + (x[1]-beta[1])**2 -beta[2]**2

# initial guess for parameters
R_m = calc_R(x_m, y_m).mean()
beta0 = [ x_m, y_m, R_m]

d = sqrt((x-x_m)**2+(y-y_m)**2)
error = R_m/sqrt(mean((d-R_m)**2))
# print(error)


# for implicit function :
#       data.x contains both coordinates of the points (data.x = [x, y])
#       data.y is the dimensionality of the response
lsc_data  = odr.Data(row_stack([x, y]),y=1,we=error,wd=error)
lsc_model = odr.Model(f_3, implicit=True)
lsc_odr   = odr.ODR(lsc_data, lsc_model, beta0)
lsc_out   = lsc_odr.run()

xc_3, yc_3, R_3 = lsc_out.beta
Ri_3 = calc_R(xc_3, yc_3)
#    residu_3 = sum((Ri_3 - R_3)**2)
lsc_out.beta
x0 = lsc_out.beta[0]
y0 = lsc_out.beta[1]
r0 = lsc_out.beta[2]

a = linspace(0,2*pi,1000)
x = r0 + r0 *cos(a)
y =  r0 *sin(a)
Angle = arctan2(y0,x0)

#Y_1=(Y-complex(x0,y0))*exp(-1j*Angle)+sqrt(x0**2+y0**2)
Y_1=(Y*exp(-1j*Angle)) # sqrt(x0**2+y0**2)
```
